```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Replace 'your_dataset.csv' with the actual path to your dataset
data = pd.read_csv('deliveries.csv')

# Data Cleaning (You need to customize this based on your dataset)
# For example, handling missing values, data type conversions, etc.

# Team Success Analysis
team_wins = data.groupby(['batting_team', 'match_id']).tail(1)
team_wins = team_wins['batting_team'].value_counts()

# Player Performance Analysis
batsman_runs = data.groupby('batsman')['batsman_runs'].sum().nlargest(10)
bowler_wickets = data[data['dismissal_kind'].notna()].groupby('bowler')['player_dismissed'].count().nlargest(10)

# Factors Contributing to Wins and Losses
extra_runs = data.groupby('batting_team')['extra_runs'].sum()
dismissal_types = data[data['dismissal_kind'].notna()]['dismissal_kind'].value_counts()

# Visualization
# Create charts to visualize team wins, top run-scorers, top wicket-takers, and other relevant data.
```

```python
# Team Success Chart
plt.figure(figsize=(10, 5))
team_wins.plot(kind='bar', color='skyblue')
plt.title('Team Success - Number of Wins')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Batsmen Chart
plt.figure(figsize=(10, 5))
batsman_runs.plot(kind='bar', color='lightcoral')
plt.title('Top Batsmen - Runs Scored')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Bowlers Chart
plt.figure(figsize=(10, 5))
bowler_wickets.plot(kind='bar', color='lightgreen')
plt.title('Top Bowlers - Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)
plt.show()
```

```python
# Factors Contributing to Wins and Losses
# Create charts to visualize factors like extra runs and dismissal types

# Extra Runs by Team Chart
plt.figure(figsize=(10, 5))
extra_runs.plot(kind='bar', color='lightblue')
plt.title('Extra Runs Conceded by Teams')
plt.xlabel('Team')
plt.ylabel('Extra Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Dismissal Types Chart
plt.figure(figsize=(10, 5))
dismissal_types.plot(kind='bar', color='lightcoral')
plt.title('Dismissal Types')
plt.xlabel('Dismissal Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

plt.show()
```
