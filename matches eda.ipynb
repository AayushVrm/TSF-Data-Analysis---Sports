```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("matches.csv")

# Display the first few rows of the dataset
print(data.head())

# Check data types and non-null values
print(data.info())

# Summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Unique values in each column
for column in data.columns:
    unique_values = data[column].unique()
    if len(unique_values) <= 10:
        print(f"Unique values in {column}: {unique_values}")
    else:
        print(f"Unique values in {column}: {len(unique_values)} unique values")

# Data Cleaning
# 1. Convert the 'date' column to datetime
data['date'] = pd.to_datetime(data['date'])

# Visualizations
# 1. Bar chart of matches played per season
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=data)
plt.xticks(rotation=90)
plt.title('Matches Played per Season')
plt.show()

# 2. Toss decision distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='toss_decision', data=data)
plt.title('Toss Decision Distribution')
plt.show()

# 3. Distribution of match results
plt.figure(figsize=(8, 6))
sns.countplot(x='result', data=data)
plt.xticks(rotation=90)
plt.title('Match Result Distribution')
plt.show()

# 4. Box plot of win by runs and win by wickets
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['win_by_runs', 'win_by_wickets']])
plt.title('Distribution of Wins by Runs and Wickets')
plt.show()

# 5. Correlation heatmap
plt.figure(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Save the analyzed data
data.to_csv("analyzed_cricket_data.csv", index=False)
```

```python
import pandas as pd

# Load the analyzed data
analyzed_data = pd.read_csv("analyzed_cricket_data.csv")
```

```python
# Count the number of wins by each team
team_wins = analyzed_data['winner'].value_counts()

# Plot a pie chart for the most successful teams
plt.figure(figsize=(8, 8))
plt.pie(team_wins, labels=team_wins.index, autopct='%1.1f%%', startangle=140)
plt.title('Most Successful Teams')
plt.show()
```

```python
# Count the number of times each player received the "Player of the Match" award
player_of_match = analyzed_data['player_of_match'].value_counts().head(15)

# Plot a pie chart for the most successful players
plt.figure(figsize=(8, 8))
plt.pie(player_of_match, labels=player_of_match.index, autopct='%1.1f%%', startangle=140)
plt.title('Most Successful Players (Player of the Match)')
plt.show()
```

```python
# Count the different results
result_counts = analyzed_data['result'].value_counts()

# Plot a pie chart for match results
plt.figure(figsize=(8, 8))
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Match Results')
plt.show()
```
